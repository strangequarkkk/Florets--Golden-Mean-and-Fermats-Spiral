```python
def polar2cart(r,theta):
    '''
This function converts plane polar coordinates(r,theta) to cartesian coordinates (x,y),using the formula :-

x=rcostheta
y=rsintheta


Note: np.cos is the mathematical function Cos(theta)
np.sin is the mathematical function Sin(theta)
theta is an angle supplied in degrees
    '''
    x=r*np.cos(theta)
    y=r*np.sin(theta)
    return x,y
```

```python
polar2cart(2,90)         #Function calculated for scalar arguments
```

```python
polar2cart(np.array([1,2,3]),np.array([0,45,90]))    #Function calculated for vector arguments
```

```python
plt.figure(figsize=(10,10))

phi = (1 + np.sqrt(5))/2

def floretpositions (nmax,ndiff):                                 #function for part b defined here with input parameters 'maximum number of florets and ndiff' and output parameters xf and yf as vectors
    n = np.arange(1,nmax+1)
    rf = np.sqrt(n)
    phi = (1 + np.sqrt(5))/2
    thetaf = (2*np.pi/phi**2+ndiff)*n
    xf,yf= polar2cart(rf,thetaf)                                  #Function from Q1 called(part 2a) & #output parameters from 'polar2cart' defined to local variables xf and yf to use them for return
    return np.array([xf]),np.array([yf])                          #Output vectors xf and yf returned

xf , yf = floretpositions(20,0)                                   #function 'floret positions' called and output parameters defined to variables xf and yf
plt.plot(xf,yf,'ro')                                              #plotting the floret positions(i.e red circles)
plt.title('Fermat Spiral')

theta = np.linspace(0,50,500)
r = np.sqrt(phi**2*theta/(2*np.pi))
x,y=polar2cart(r,theta)                                           #function in Q1 called(part 2a)

plt.plot(x,y)                                                     #plotting the continuous line(spiral)




```

```python
#part c using for loop


nmax=20                                                          # total no. of florets defined
r=2                                                              # no. of rows to be used for subplots defined
c=3                                                              # no. of columns to be used for subplots defined
i=1                                                              # position index for subplots defined
fig=plt.figure(figsize=(16,8))                                   # Figure with size 16 inches x 8 inches
plt.subplots_adjust(wspace=0.4, hspace=0.6, top=0.85)

sequence=np.arange(-0.25,0.26,0.1)                               #range of ndiff defined in sequence

for ndiff in sequence:
    xf1,yf1=floretpositions(nmax,ndiff)
    plt.subplot(r, c, i)
    plt.plot(xf1,yf1,'ro')
    plt.title('20 floret positions with ndiff = %.2f ' % ndiff)
    i= 1 + i                                                     #index updated from position 1(top left) to go to position 6(bottom right)
```

```python

#Part c using while loop

nmax2=20                                                         # total no. of florets defined
ndiff2=-0.25
r=2                                                              # no. of rows to be used for subplots defined
c=3                                                              # no. of columns to be used for subplots defined
i=1                                                              # position index for subplots defined
fig=plt.figure(figsize=(16,8))                                   # Figure with size 16 inches x 8 inches
plt.subplots_adjust(wspace=0.4, hspace=0.6, top=0.85)

while ndiff2 < 0.26:                                             #loop condition 
    xf2,yf2=floretpositions(nmax2,ndiff2)
    plt.subplot(r, c, i)
    plt.plot(xf2,yf2,'bo')
    plt.title('20 floret positions with ndiff = %.2f ' % ndiff2)
    ndiff2 = ndiff2 + 0.1
    i= i + 1                                                     #index updated from position 1(top left) to go to position 6(bottom right)

```
